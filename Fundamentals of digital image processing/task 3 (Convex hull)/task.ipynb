```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def thickening(image, element, iterations=1):
    thickened = image.copy()
    for _ in range(iterations):
        thickened = cv2.dilate(thickened, element)

    return thickened
```

```python
def convex_hull(image, element):
    ret, image = cv2.threshold(image, 127, 255, 0)
    convex_hull_image = thickening(image, element, iterations = 10)
    return convex_hull_image
```

```python
image = cv2.imread("image1.png", cv2.IMREAD_GRAYSCALE)
kernel = np.ones((3, 3), dtype=np.uint8)
convex_hull = convex_hull(image, kernel)

fig, axes = plt.subplots(1,  2, figsize=(15,  5))

axes[0].imshow(image, cmap=plt.cm.gray)
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].imshow(convex_hull, cmap=plt.cm.gray)
axes[1].set_title('Convex Hull')
axes[1].axis('off')

plt.tight_layout()
plt.show()
```
