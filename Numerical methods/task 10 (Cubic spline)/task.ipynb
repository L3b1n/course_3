```python
import numpy as np
import scipy.interpolate as interpolate

# Узлы интерполяции
x = np.array([0, 0.333333, 0.666667, 1])

# Значения функции в узлах
y = np.array([0, 2.22222, 2.22222, 0])

# Вторая производная в крайних точках
fpp_0 = fpp_3 = -20

# Создание кубического сплайна
cubic_spline = interpolate.CubicSpline(x, y, bc_type=((1, fpp_0), (1, fpp_3)))

# Вычисление значений сплайна в заданных точках
x_new = np.linspace(0, 1, 100)
y_new = cubic_spline(x_new)

# Визуализация сплайна
import matplotlib.pyplot as plt

plt.plot(x, y, 'o', label='Исходные точки')
plt.plot(x_new, y_new, '-', label='Кубический сплайн')
plt.legend()
plt.show()
```
