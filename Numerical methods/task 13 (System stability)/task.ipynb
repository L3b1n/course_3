```python
import numpy as np
import pandas as pd
import scipy.sparse as sp
from tqdm.notebook import tqdm
import ast
import scipy.stats as ss
import matplotlib.pyplot as plt
```

```python
h = 0.1
r = 0.19
a = 0.5
m = int(2 / h) + 1
n = int(2 / (a * r * h)) + 1
print(n, m)

x0 = -1
t0 = 0

dp = np.ones((n, m))
F = lambda x: np.sin(x)


for i in range(n):
    for j in range(m):
        dp[i, j] = F(a * i * r * h - (x0 + j * h))

for i in range(1, n):
    for j in range(1, m):
        dp[i, j] = (dp[i-1, j] + (a * r * dp[i, j - 1])) / (a*r+1)

x = [-1]
for j in range(m - 1):
    x.append(x[-1] + h)

y_correct = []
for j in range (0, m):
    y_correct.append(F(a*(n-1)*r*h - (x0 + j*h)))


figure, (plot1, plot2) = plt.subplots(1, 2)
plot1.plot(x, dp[-1, :])
plot1.set_title("Actual result")
plot2.plot(x, y_correct)
plot2.set_title("Correct result")
plt.show()
```
